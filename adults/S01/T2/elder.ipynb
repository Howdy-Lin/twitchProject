```python
# 

results = []  # 

for i in range(1, 6):
    file_name = f"rd{i}.csv"
    df = pd.read_csv(file_name, header=1)
    
    # DataFrameHeight
    df_height = df["Height (cm)"]
    
    # Height15
    T1 = df_height.gt(15).idxmax()
    
    # Grip ForceT1100
    subset = df["Grip Force(g)"][T1:T1+100]
    grip_force_data = list(df["Grip Force(g)"])
    
    # subset
    max_value = max(subset)
    max_index = grip_force_data.index(max_value) + 1
    
    # T2-T1
    delta_T = max_index - T1
    
    # 
    results.append((delta_T, max_value))

# DataFrame
results_df = pd.DataFrame(results, columns=["T2-T1", "Max Value"])

# Excel
results_df.to_csv("results.csv", index=False)
```
